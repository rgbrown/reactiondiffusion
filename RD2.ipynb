### The Brusselator
The Brusselator is a simple PDE system that is capable of exhibiting some fairly complex behaviour. We use the following PDE description which allows for spatially varying parameters $a(x)$ and $b(x)$
\begin{align}
    \phi_t &= D_\phi \phi_{xx} + a(x) + \phi^2\psi - (b(x) + 1)\phi \\
    \psi_t &= D_\phi \phi_{xx} + b(x) \phi - \phi^2 \psi
\end{align}
We complete the problem with boundary conditions:
\begin{align}
    &\phi_x(-1, t) = \phi_x(1, t) = 0, & &\psi_x(-1, t) = \psi_x(1, t) = 0 \\
    &\phi(x, 0) = 1, & &\psi(x, 0) = f(x).
\end{align}
Let's define some functions $a(x), b(x), f(x)$, and our parameters

```python
x, phi, psi = symbols('x, phi, psi')
D_phi = 1e-3
D_psi = 0.
npts = 1001
a = 1
b = 2 - x
f = b + 0.2 * sin(4*pi*x)
```

```python
f_phi = a + phi**2 * psi - (b + 1)*phi
f_psi = b*phi - phi**2*psi
J = Matrix([[diff(f_phi, phi), diff(f_phi, psi)],
            [diff(f_psi, phi), diff(f_psi, psi)]])
Jfun = lambdify([x, phi, psi], J)
phi_fun = lambdify([x, phi, psi], f_phi)
psi_fun = lambdify([x, phi, psi], f_psi)
f_fun = lambdify((x), f, modules="numpy")
# scipy.sparse.block_diag([Jfun(1,1,x) for x in np.linspace(-1, 1)])
```

We're going to need to be able to define our finite difference matrix, this function is the ticket

```python
def diff2(npts, h=1.):
    """
    Create second derivative finite difference matrix, assuming Neumann boundary conditions
    """
    diags = [np.zeros(npts) + 1., np.zeros(npts) - 2., np.zeros(npts) + 1.]
    diags[0][1] = 2.
    diags[2][-2] = 2.
    return 1/(h**2) * scipy.sparse.dia_matrix((diags, [1, 0, -1]), shape=(npts,npts))
```

We now have all the bits and pieces we need to get the ODE set up. Let's start by forming the differential operator.

```python
x = np.linspace(-1, 1, npts) # overwrites symbolic variable, but that's OK
h = (x[-1] - x[0]) / (npts - 1)
nvars = 2
diffusion = np.array([D_phi, D_psi])

D_loc = diff2(npts, h).tocoo() # The vanilla differential operator
N = nvars * npts
D = scipy.sparse.dia_matrix((N, N))
for i in range(nvars):
    if diffusion[i] > 0:
        r, c = nvars*D_loc.row + i, nvars*D_loc.col + i
        D = D + scipy.sparse.coo_matrix((diffusion[i] * D_loc.data, (r, c)), (N, N)).todia()
```

Next, we'll set up the right hand side function and Jacobian

```python
def rhs(u, t):
    du = np.zeros(u.shape)
    phi, psi = u[::2], u[1::2]
    du[::2] = phi_fun(x, phi, psi)
    du[1::2] = psi_fun(x, phi, psi)
    return D*u + du

u0 = np.zeros(nvars * npts)
u0[::2] = 1.
u0[1::2] = f_fun(x)

def jac(u, t):
    phi, psi = u[::2], u[1::2]
    Jloc = scipy.sparse.block_diag([Jfun(xi, phi_i, psi_i) for (xi, phi_i, psi_i) in zip(x, phi, psi)])
    return (Jloc + D).todia()
    
def Dfun(u, t):
    J = jac(u, t)
    return np.flipud(J.data)

J0 = jac(u0, 0)
ml, mu = J0.offsets[0], J0.offsets[-1]
print [ml, mu]
```

```python
t = np.linspace(0, 40, 1000)
U = odeint(rhs, u0, t, Dfun=Dfun, ml=ml, mu=mu, rtol=1e-3, atol=1e-6).T
img = plt.imshow(U[::2,:], aspect="auto")
img.set_cmap('hot')
```
